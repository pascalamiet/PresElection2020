```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data_raw/gdp_us.csv',index_col=0)
df
```

```python
df['GDP'].plot(legend=True,color='black')
plt.ylabel('US Gross Domestic Product (USD)')
plt.xlabel('Time') 
plt.savefig('../data_clean/gdp_us.pdf',bbox_inches='tight')
plt.show()
```
